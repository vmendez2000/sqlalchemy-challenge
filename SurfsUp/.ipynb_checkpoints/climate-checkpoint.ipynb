```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
print(Base.classes.keys())
```

```python
# Save references to each table
Station = Base.classes.station
Measurement = Base.classes.measurement
```

```python
# Create session to interact with the database
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
most_recent_date = session.query(func.max(Measurement.date)).scalar()
print( {most_recent_date})
```

```python
# Calculate the date one year from the last date in data set.
one_year_ago = dt.datetime.strptime(most_recent_date, "%Y-%m-%d") - dt.timedelta(days=365)
print(f"Date one year ago: {one_year_ago}")

# Perform a query to retrieve the data and precipitation scores
precipitation_data = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date >= one_year_ago).\
    order_by(Measurement.date).all()

# Convert query results to a DataFrame
precip_df = pd.DataFrame(precipitation_data, columns=["date", "precipitation"])

# Drop any rows with missing values
precip_df.dropna(inplace=True)

# Sort the DataFrame by date
precip_df.sort_values("date", inplace=True)

# Plot the precipitation data
precip_df.plot(x="date", y="precipitation", rot=90, figsize=(10, 6), legend=False, 
               title="Precipitation Over the Last 12 Months", label="Precipitation")  

# Add the legend
plt.legend()

# Label the axes
plt.ylabel("Precipitation (inches)")
plt.xlabel("Date")

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate summary statistics for the precipitation data
precip_stats = precip_df['precipitation'].describe()

# Display the summary statistics
print("Precipitation")
precip_stats
```

# Exploratory Station Analysis

```python
# Query to calculate the total number of stations
total_stations = session.query(func.count(Station.station)).scalar()
total_stations
```

```python
# Query to find the most active stations, ordered by observation count
active_stations = session.query(
    Measurement.station, func.count(Measurement.id).label('observation_count')
).group_by(Measurement.station).order_by(func.count(Measurement.id).desc()).all()

# Display the stations and their counts
print("Most Active Stations (Ordered by Observation Count):")
for station, count in active_stations:
          print(f"Station {station}: {count} observations")
```

```python
# Query to calculate the lowest, highest, and average temperature for the most active station
temperature_stats = session.query(
    func.min(Measurement.tobs).label('min_temp'),
    func.max(Measurement.tobs).label('max_temp'),
    func.avg(Measurement.tobs).label('avg_temp')
).filter(Measurement.station == 'USC00519281').all()

# Display the results
temperature_stats
```

```python
# Set the most recent date 
most_recent_date = '2017-08-23'
# Calculate the date one year ago from the most recent date
one_year_ago = dt.datetime.strptime(most_recent_date, "%Y-%m-%d") - dt.timedelta(days=365)

# Query for the last 12 months of temperature observations based on the found most recent date
tobs_data = session.query(Measurement.tobs).filter(
    Measurement.station == 'USC00519281',
    Measurement.date >= one_year_ago
).all()

# Extract temperatures from the query result
temperatures = [t[0] for t in tobs_data]

# Plot the histogram with a label for the legend
plt.figure(figsize=(10, 6))
plt.hist(temperatures, bins=12, edgecolor='black', label='tobs')  
plt.title(f"Temperature Observations for Station USC00519281, Last 12 Months")
plt.xlabel("Temperature (F)")
plt.ylabel("Frequency")

# Add the legend
plt.legend()

# Display the plot
plt.show()
```

# Close Session

```python
# Close Session
session.close()
```
